```python
from pyiceberg.catalog import load_catalog
# Load the REST catalog
catalog = load_catalog('default')
# Load the users table
tbl = catalog.load_table('accounts.users')
# Create a scan with filter on last_name
sc = tbl.scan(row_filter="last_name LIKE 'M%' ")
# Convert the scan results into a Pandas dataframe
df = sc.to_arrow().to_pandas()
df
```
