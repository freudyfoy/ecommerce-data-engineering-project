```python
%%sql
CREATE DATABASE IF NOT EXISTS accounts;
```

```python
%%sql
CREATE TABLE IF NOT EXISTS accounts.users (
id INT,
first_name VARCHAR(25),
last_name VARCHAR(25),
email VARCHAR(50)
)
USING iceberg
PARTITIONED BY (truncate(1, last_name))
```
